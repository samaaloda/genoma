```python
import pandas as pd
og_cosmic_db = pd.read_csv("Cosmic_CompleteCNA_v102_GRCh37.tsv", sep="\t")
og_cosmic_db 
```

```python
#The COSMIC_PHENOTYPE_ID in the cosmic complete database matches with the COSMIC_PHENOTYPE_ID in the classification database
classif_db = pd.read_csv("Cosmic_Classification_v102_GRCh37.tsv", sep="\t")
classif_db
```

```python
cosmic_db = og_cosmic_db #og wont be changed
cosmic_db["MUT_TYPE"] = cosmic_db["MUT_TYPE"].astype(str).str.lower().map({"gain": 1, "loss": 0})
cosmic_db = cosmic_db.dropna()
cosmic_db = cosmic_db.sample(n=1000, random_state=42) #making the randommness reproducible
```

```python
X = cosmic_db[["TOTAL_CN","MINOR_ALLELE","MUT_TYPE", "GENOME_START", "GENOME_STOP"]]
X[~pd.isna(X["MINOR_ALLELE"])] #dropping MINOR_ALLELE since they are all NaN
X = X.drop(columns=["MINOR_ALLELE"])
X= X.to_numpy() #must be in array form for sklearn
X
```

```python
Y=[]

for phen_id in cosmic_db["COSMIC_PHENOTYPE_ID"]:
    try:
        arr = classif_db[classif_db["COSMIC_PHENOTYPE_ID"] == phen_id]
        #print(arr["PRIMARY_HISTOLOGY"])
        #Y.append(arr.iloc[0])
        Y.append(arr["PRIMARY_HISTOLOGY"].values[0]) 
    except :
        print(f"{phen_id} not found in classification database.")

Y
```

```python
clf = RandomForestClassifier()
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
clf.fit(X_train, Y_train)
print(clf.predict(X_test))
print(Y_test)
print(clf.score(X_test, Y_test))
```
